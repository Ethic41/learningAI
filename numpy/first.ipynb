```python
# Sample Plot with Jupyter in VSCode
import matplotlib.pyplot as plt

def mendelbrot(h, w, maxit=20):
    y,x = np.ogrid[ -1.4:1.4:h*1j, -2:0.8:w*1j ]
    c = x+y*1j
    z = c
    
    divtime = maxit + np.zeros(z.shape, dtype=int)
    
    for i in range(maxit):
        z = z ** 2 + c
        diverge = z*np.conj(z) > 2**2
        div_now = diverge & (divtime==maxit)
        divtime[div_now] = i
        z[diverge] = 2

    return divtime

figure = plt.imshow(mendelbrot(400,400))
```

```python
# ok so this should be easy right
# lets generate an 8x8 array
a = np.ones([8,8])
# let's flatten it for some
# slicing tricks
a = a.flatten() 
print(a)
```

```python
# ok next we let's fill
# even indices with '0's
# Beware: slicing Ahead
a[::2] = 0
print(a)
```

```python
# all good right?
# now let's build up
# our dimension
a = a.reshape([8,8])
print(a)
```

```python
# oops that doesn't look
# like a checkerboard ryt?
# now heres what I found,
# let's repeat everything
# but for 9x9
b = np.ones([9,9]).flatten()
b[::2] = 0
b = b.reshape([9,9])
print(b)
```

```python
# yeah u figured! for odd numbers
# it works but for even we're 
# screwed! or maybe not :)
# turns out we could get 8x8
# out of 9x9, remove one row
# one column :), this is how:)
x = b  # from above
# let's remove the first row
# to obtain 9x8
x = x[1:]
print(x)
```

```python
# now let's reshape the
# the chick a lil bit :)
x = x.reshape([8,9])
# and transpose her a bit
x = x.T
# let's remove one column
# :) or is it a row :>
x = x[1:]
# Voila!
print(x)
```

```python
from matplotlib import pyplot as plt

x = np.linspace(-np.pi, np.pi, 256, endpoint=True)
cosine, sine = np.cos(x), np.sin(x)

plt.plot(x, cosine)
plt.plot(x, sine)

plt.show()
```

```python
plt.figure(figsize=(8,6), dpi=80)
plt.subplot(1,1,1)
plt.plot(x, cosine, color="blue", linewidth=1.0, linestyle="-")
plt.plot(x,sine, color="green", linewidth=1.0, linestyle="-")
plt.xlim(-4.0, 4.0)
plt.xticks(np.linspace(-4,4,9, endpoint=True))
plt.ylim(-1.0,1.0)
plt.yticks(np.linspace(-1,1,8, endpoint=True))
plt.show()
```

```python
# control figure size
plt.figure(figsize=(8,6), dpi=80)
# creating subplots
plt.subplot(1,1,1)
# plotting data
plt.plot(x, cosine, color="blue", linewidth=1.7, linestyle="-", label="cosine")
plt.plot(x, sine, color="green", linewidth=1.7, linestyle="-", label="sine")
# creating a legend
plt.legend(loc="upper left")
# setting limit for the axis
plt.xlim(x.min() * 1.1, x.max() * 1.1)
plt.ylim(cosine.min() * 1.1, cosine.max() * 1.1)
# setting ticks for the axis
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])
plt.yticks([-1, 0, 1],[r'$-1$', r'$0$', r'$+1$'])
ax = plt.gca()  # get current axis
# creating spines
ax.spines["right"].set_color('none')
ax.spines["top"].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data', 0))
# Annotation cosine
t = 2 * np.pi / 3
plt.plot([t, t], [0, np.cos(t)], color="blue", linewidth=2.5, linestyle="--")
plt.scatter([t, ], [np.cos(t), ], 50, color="blue")
plt.annotate(r'$cos(\frac{2\pi}{3})=-\frac{1}{2}$', xy=(t, np.cos(t)), xycoords='data', xytext=(-90, -50),
textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))

# Annotation sine
plt.plot([t, t], [0, np.sin(t)], color="red", linewidth=2.5, linestyle="--")
plt.scatter([t, ], [np.sin(t), ], 50, color="red")
plt.annotate(r'$sin(\frac{2\pi}{3})=\frac{\sqrt{3}}{2}$', xy=(t, np.sin(t)), xycoords='data', xytext=(+10, +30),
textcoords='offset points', fontsize=16, arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
# display all the graph
plt.show()
```
